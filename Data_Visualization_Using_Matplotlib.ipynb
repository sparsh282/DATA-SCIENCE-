```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (3,2))

x = np.linspace(0,5,11)
y = x**2

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.12,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')

axes2.plot(y,x)
axes2.set_title('Smaller plot')

plt.tight_layout()
```

```python
#plt.subplots automatically adds the axes to your figure so no need to add stuff

#Unpacking a Tuple
fig, axes = plt.subplots(nrows=1, ncols=2,figsize = (5,2),dpi = 200)

#Fixes the layout problem
plt.tight_layout()

#Different plotting Paramteres that can be added
axes[0].plot(x,y,color='orange',linewidth = 2, alpha = 0.5,linestyle = '--',marker = 'o',markersize = '10', markerfacecolor = 'yellow',label = 'X')
axes[0].plot(y,x,color='yellow',linewidth = 2 ,alpha = 0.5,linestyle = '--',marker = 'o',markersize = '10', markerfacecolor = 'yellow',label = 'Y')
axes[0].set_title('1stplot')
axes[0].set_ylabel('Yo')
axes[0].set_xlabel('yo yo')
axes[0].legend(loc=0)
axes[0].set_xlim([0,1])
axes[0].set_ylim([0,2])

axes[1].plot(x,y)
axes[1].set_title('2ndplot')

#Save figure
fig.savefig('my_picture.jpg')
```

```python
#For loop way of adding the axes
for current_ax in axes:
  current_ax.plot(x,y)
```
